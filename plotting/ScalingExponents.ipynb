```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
from __future__ import print_function
import pandas as pd
import numpy as np
import scipy
from matplotlib import pyplot as plt, gridspec
from plotutils import *

basedir = '../benchmarks_py/sfi_scaling006'

DO_SAVE            = True # whether to save pdf output
```

```python
PLOT_COLS = ['Comprehension', 'MeanGrammarDrift', 'AgentGrammarDrift', 'GrammarVar']
PLOT_SCALING_TYPES = ['agents','memes','lexes','lexesmemes','lexesmemesagents']
fig=plt.figure(figsize=(15,12))
gs = gridspec.GridSpec(len(PLOT_COLS),len(PLOT_SCALING_TYPES))
#    plt.suptitle('Scaling with %s'%desc, fontsize=17, y=1.01)

for cndx, scalingtype in enumerate(PLOT_SCALING_TYPES):
    COL, desc, data, mutators = get_data(basedir, scalingtype)
    smutators = sorted(mutators)

    for pndx, pCol in enumerate(PLOT_COLS):
        print("Doing %s" % pCol)

        for mndx, m in enumerate(smutators):
            _, xs, ys, ylabel = get_scaling_vals(data, COL, m, pCol)
            cTitle = 'Measure: %s' % COL_NAMES.get(pCol, pCol)
            if len(xs):
                plt.subplot(gs[pndx, cndx])
                slope, _ = plot_scatter(xs, ys, loglog=True, regression_line=True)
                if cndx == 0:
                    plt.ylabel(ylabel)
                if pndx == len(PLOT_COLS) - 1:
                    plt.xlabel(desc)
                if pndx == 0:
                    cTitle ='Scaling with %s\n'%desc + cTitle
            plt.title(cTitle)
handles, labels = plt.gca().get_legend_handles_labels()
leg=fig.legend(handles, smutators, loc='upper center', ncol=len(smutators), fontsize=13,bbox_to_anchor=(0., 0.93, 1., .102))
plt.tight_layout()

if DO_SAVE:
    plt.savefig('output/scalingexponents2019.pdf', bbox_extra_artists=(leg,), bbox_inches='tight')
```
